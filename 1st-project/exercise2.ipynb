```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv('data.csv')
data
```

```python
#calculate the functions of B and C for A 
# C
data['logC'] =1.5*np.log(data.A)

# B
data['loglogB'] =data.A**0.51
data
```

```python
plt.figure(dpi = 130)
plt.scatter(x=data.A,y=data.B,s=1)
plt.scatter(x=data.A,y=data.loglogB,s=1)
plt.xscale('log')
plt.yscale('log')
plt.title('function B of A is b = a^0.51')


plt.figure(dpi = 130)
plt.scatter(x=data.A,y=data.C,s=1)
plt.scatter(x=data.A,y=data.logC,s=1)
plt.xscale('log')
plt.title('function C of A is c =1.5*log(A)')
```

### Συμπεράσματα

Η συνάρτηση B(Α) έχει γραμμική απεικόνιση όταν και οι δύο άξονες είναι λογαριθμικοί, άρα η συνάρτηση έχει τη γενική μορφή $y = w*x^k$, *w,k παράμετροι*.

Η συνάρτηση C(Α) έχει γραμμική απεικόνιση όταν μόνο ο άξονας x είναι λογαριθμικός, οπότε η συνάρτηση θα είναι της μορφής $y = w*log(x)$, *w παράμετρος*.

Η Β συνάρτηση είναι η $b = a^{0.51}$ και η C η $c =1.5*log(a)$.
